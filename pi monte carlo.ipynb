```python
import random
import matplotlib.pyplot as plt # type: ignore
```

```python
c_pi = []
n = int(input("Number of points: "))
circle = 0

for i in range(n):
    x, y = random.random(), random.random()
    distance = (x**2 + y**2)**0.5 
    if distance < 1:
        circle = circle + 1

    c_pi.append(4 * circle/(i + 1))
print(c_pi[-1])
```

```python
plt.plot(c_pi)
```
